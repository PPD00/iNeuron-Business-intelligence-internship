```python
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
data['DateKey'] = pd.to_datetime(data['DateKey'])
```

```python
data['DateKey']
```

```python
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])
```

```python
data['Invoice_Year'] = data['Invoice Date'].dt.year
data['Invoice_Month'] = data['Invoice Date'].dt.month
data['Invoice_Quarter'] = data['Invoice Date'].dt.quarter
data['Invoice_Day'] = data['Invoice Date'].dt.day
```

```python
data.info()
```

```python
## Creating DataFrame only with necessary values.
## Considering the data which is realted to sales.

sales_data = data[['CustKey','Item','Invoice Date','Invoice_Year','Invoice_Quarter','Invoice_Month','Invoice_Day',
                          'Sales Quantity','Sales Amount','Sales Amount Based on List Price','Discount Amount','Sales Margin Amount',
                          'Sales Cost Amount','Sales Rep','U/M','List Price','Sales Price']]
```

```python
sales_data.isnull().sum()
```

```python
sales_data.shape
```

```python
for col in sales_data.columns:
    print(f"{col}: {sales_data[col].nunique()} unique values")

## Data type conversion
sales_data['Invoice Date'] = pd.to_datetime(sales_data['Invoice Date'])  # Convert to datetime
sales_data['Invoice_Month'] = pd.Categorical(sales_data['Invoice_Month'], ordered=True)
```

## Univariate Analysis

```python
# Sales Amount Distribution
plt.figure(figsize=(8,5))
sns.histplot(sales_data['Sales Amount'], bins=50, kde=True)
plt.title('Sales Amount Distribution')
plt.show()

# Sales Quantity Distribution
plt.figure(figsize=(8,5))
sns.histplot(sales_data['Sales Quantity'], bins=50, kde=True)
plt.title('Sales Quantity Distribution')
plt.show()
```

## Bivariate Analysis

```python
# Sales Amount vs. Discount Amount
plt.figure(figsize=(8,5))
sns.scatterplot(x=sales_data['Discount Amount'], y=sales_data['Sales Amount'])
plt.title('Sales Amount vs. Discount Amount')
plt.show()

# Sales Margin Amount vs. Sales Quantity
plt.figure(figsize=(8,5))
sns.scatterplot(x=sales_data['Sales Quantity'], y=sales_data['Sales Margin Amount'])
plt.title('Sales Margin Amount vs. Sales Quantity')
plt.show()
```

 - 

## Time Series Analysis

```python
sales_data['Invoice Date'] = pd.to_datetime(sales_data['Invoice Date'])
sales_data['Month'] = sales_data['Invoice Date'].dt.to_period('M')

# Monthly Sales Trend
monthly_sales = sales_data.groupby('Month')['Sales Amount'].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.grid()
plt.show()
```

```python
## monthly sales pattern

monthly_sales = sales_data.groupby('Invoice_Month')['Sales Amount'].sum()
monthly_sales.plot(kind='line', marker='v', title='Monthly Sales')
plt.xlabel('Month')
plt.show()
```

## Top Products and Sales Representation

```python
## Top products
top_products = sales_data.groupby('Item')['Sales Amount'].sum().nlargest(10)
top_products.plot(kind='bar', title='Top Products by Sales')
plt.ylabel('Sales Amount')
plt.show()
```

```python
sales_by_unit = sales_data.groupby('U/M')['Sales Amount'].sum()
sales_by_unit.plot(kind='pie', autopct='%1.1f%%', title='Sales by Unit of Measure')
plt.ylabel('')
plt.show()
```

```python
sales_by_rep = sales_data.groupby('Sales Rep')['Sales Amount'].sum().nlargest(10)
sales_by_rep.plot(kind='barh', title='Sales by Sales Rep', color='skyblue')
plt.xlabel('Sales Amount')
plt.show()
```

```python
# Sales Amount by Sales Rep
plt.figure(figsize=(20,6))
sns.barplot(x='Sales Rep', y='Sales Amount', data=sales_data, estimator=sum)
plt.xticks(rotation=45)
plt.title('Total Sales Amount by Sales Rep')
plt.show()
```

## Outlier Detection Analysis

```python
sns.boxplot(x=sales_data['Sales Amount'])
plt.title('Outlier Analysis for Sales Amount')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=sales_data['Sales Margin Amount'])
plt.title('Outliers in Sales Margin Amount')
plt.show()
```

```python
sales_data['Margin Percentage'] = (sales_data['Sales Margin Amount'] / sales_data['Sales Amount']) * 100

# Distribution of Margin Percentage
sns.boxplot(x='Invoice_Year', y='Margin Percentage', data=sales_data)
plt.title('Margin Percentage by Year')
plt.show()
```

## Correlation

```python
# Correlation
correlation_matrix = sales_data[['Sales Quantity', 'Sales Amount', 'Discount Amount', 'Sales Margin Amount', 'Sales Cost Amount']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
from scipy.stats import ttest_ind

## comparing average sales with and without discounts
discounted = sales_data[sales_data['Discount Amount'] > 0]['Sales Amount']
non_discounted = sales_data[sales_data['Discount Amount'] == 0]['Sales Amount']

t_stat, p_value = ttest_ind(discounted, non_discounted)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
